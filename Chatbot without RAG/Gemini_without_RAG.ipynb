```python
import google.generativeai as genai
import os
import sys
```

```python
genai.configure(api_key="YOUR_API_KEYS")
llm1 = genai.GenerativeModel('gemini-1.5-flash', system_instruction="คุณเป็น AI assistant ที่เป็นมิตรและช่วยเหลือ")
```

```python
def Normal_LLM(query, llm1, chat_history):

    full_query = "\n".join([f"User: {q}\nAssistance: {a}" for q, a in chat_history])
    full_query += f"\nUser: {query}"

    chain1 = llm1.generate_content(f"{full_query}")

    result = chain1.text

    return result
```

```python
chat_history = []
query = None  # Initialize query to avoid potential reference error

while True:
    if not query:
        query = input("User: ")
    if query in ['quit', 'q', 'exit']:
        break
    result = Normal_LLM(query, llm1, chat_history)
    print("Chatbot:", result)

    chat_history.append((query, result))
    query = None
```
